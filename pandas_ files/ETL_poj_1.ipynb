```python
### Create database Vehicles
```

```python
Vehicles_file="Resources/Motor_Vehicle_Collisions_-_Vehicles.csv"
NW_motor_col=pd.read_csv(Vehicles_file)
NW_motor_col
```

```python

# Create a filtered dataframe from specific columns
chosen_cols=["COLLISION_ID","CRASH_DATE",'CONTRIBUTING_FACTOR_1','CONTRIBUTING_FACTOR_2']
collision_col_1=NW_motor_col[chosen_cols].copy()
# Rename the column headers
collision_col_1 = collision_col_1.rename(columns={ "COLLISION_ID":"id",
                                                          "CRASH_DATE": "Crash_date",
                                                          "CONTRIBUTING_FACTOR_1": "Veh_Factor_One",
                                                          "CONTRIBUTING_FACTOR_2": " Veh_Factor_Two"                   
                                                          })
collision_col_1
```

```python
# Create a filtered dataframe from specific columns
chosen_col=['VEHICLE_ID','VEHICLE_MAKE','VEHICLE_YEAR','STATE_REGISTRATION','VEHICLE_TYPE']
NW_motor_col=NW_motor_col[chosen_col].copy()
# Rename the column headers
NW_motor_col = NW_motor_col.rename(columns={"VEHICLE_ID": "id",
                                                          "VEHICLE_MAKE":"Make",
                                                          "VEHICLE_YEAR": "Vehicle_year",
                                                          "STATE_REGISTRATION": "State_registration",  
                                                          "VEHICLE_TYPE": "Type"
                                                        
                                                          })
# "DRIVER_SEX": "Gender",
# "CRASH_DATE": "Crash_date",
# # Clean the data by dropping duplicates and setting the index
# NW_motor_col.drop_duplicates("id", inplace=True)
# NW_motor_col.set_index("id", inplace=True)
NW_motor_col.head()
```

```python
### Create database Person
```

```python
Person_file="Resources/Motor_Vehicle_Collisions_-_Person.csv"
NW_person_col=pd.read_csv(Person_file)
NW_person_col
```

```python
# Create a filtered dataframe from specific columns
chosen_cols=[ "COLLISION_ID","CRASH_DATE",'CONTRIBUTING_FACTOR_1','CONTRIBUTING_FACTOR_2']
collision_col_2=NW_person_col[chosen_cols].copy()
# Rename the column headers
collision_col_2 = collision_col_2.rename(columns={        "COLLISION_ID":"id",
                                                          "CRASH_DATE": "Crash_date",
                                                          "CONTRIBUTING_FACTOR_1": " Ped_Factor_One",
                                                          "CONTRIBUTING_FACTOR_2": " Ped_Factor_Two"
                                                          })
collision_col_2
```

```python
# Create a filtered dataframe from specific columns
chosen_cols=['PERSON_ID','PERSON_SEX','PERSON_AGE','PERSON_INJURY','PED_ROLE']
NW_person_col=NW_person_col[chosen_cols].copy()
# Rename the column headers
NW_person_col = NW_person_col.rename(columns={"PERSON_ID": "id",
                                                          "PERSON_SEX": "Gender",
                                                          "PERSON_AGE": "Age",
                                                          "PERSON_INJURY": "Person_injured",
                                                          "PED_ROLE": "Pedestrian_role"                       
                                                          })
# Clean the data by dropping duplicates and setting the index
# NW_person_col.drop_duplicates("id", inplace=True)
# NW_person_col.drop_duplicates("Vehicle_id", inplace=True)
# NW_person_col.set_index("id", inplace=True)

NW_person_col.head()
```

```python
# Concat collision_col_1 & collision_col_2
Collision_df=pd.concat([collision_col_1, collision_col_2])
# Clean the data by dropping duplicates and setting the index
Collision_df.drop_duplicates("id", inplace=True)
Collision_df.set_index("id", inplace=True)
Collision_df
```
