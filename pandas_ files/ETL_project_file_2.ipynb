```python
import pandas as pd 
from sqlalchemy import create_engine
import psycopg2
```

```python
## collision table 
```

```python
Collision_file="Resources/collision.csv"
Collision_df=pd.read_csv(Collision_file)

# Clean the data by dropping duplicates and setting the index
Collision_df.drop_duplicates("id", inplace=True)
Collision_df.set_index("id", inplace=True)
Collision_df
```

```python
##Vehicle Table
```

```python
Vehicles_file="Resources/Vehicle.csv"
Vehicles_df=pd.read_csv(Vehicles_file)
Vehicles_df=Vehicles_df.dropna(subset=["id"],how='all')
# Clean the data by dropping duplicates and setting the index

Vehicles_df.drop_duplicates("id", inplace=True)
Vehicles_df.set_index("id", inplace=True)

# Cleaning 
chosen_cols=['Make','Vehicle_year','State_registration','Type']
Vehicles_df= Vehicles_df[chosen_cols].copy()
Vehicles_df
```

```python
# Person Table 
```

```python
Person_file="Resources/person.csv"
Person_df=pd.read_csv(Person_file)
Person_df=Person_df.dropna(subset=["id"],how='all')
# Clean the data by dropping duplicates and setting the index
Person_df.drop_duplicates("id", inplace=True)
Person_df.set_index("id", inplace=True)
# Cleaning 
Person_df=Person_df[['Gender','Age','Person_injured','Pedestrian_role']]

Person_df
```

```python
# Veh_collision Table
```

```python
Veh_collision_file="Resources/Veh_collision.csv"
Veh_collision_df=pd.read_csv(Veh_collision_file)
Veh_collision_df=Veh_collision_df.rename(columns={'Collison_id':'Collision_id'})
# Clean the data by dropping duplicates and setting the index
Veh_collision_df.drop_duplicates("Vehicle_id", inplace=True)
Veh_collision_df.drop_duplicates("Collision_id", inplace=True)
Veh_collision_df=Veh_collision_df[['Collision_id','Vehicle_id','Damage']]
Veh_collision_df
```

```python
# per_collision Table
```

```python
per_collision_file="Resources/per_collision.csv"
per_collision_df =pd.read_csv(per_collision_file)
per_collision_df=per_collision_df.rename(columns={'Collison_id':'Collision_id'})
# Clean the data by dropping duplicates and setting the index
per_collision_df.drop_duplicates("Person_id", inplace=True)
per_collision_df.drop_duplicates("Collision_id", inplace=True)
per_collision_df=per_collision_df[['Collision_id','Person_id','Complaint']]
per_collision_df
```

```python
# per_vehicle Table
```

```python
per_vehicle_file="Resources/per_vehicle.csv"
per_vehicle_df=pd.read_csv(per_vehicle_file)
# Clean the data by dropping duplicates and setting the index
per_vehicle_df.drop_duplicates("Person_id", inplace=True)
per_vehicle_df.drop_duplicates("Vehicle_id", inplace=True)
per_vehicle_df=per_vehicle_df[['Person_id','Vehicle_id','Driver_sex','Driver_license_status']]
per_vehicle_df
```
